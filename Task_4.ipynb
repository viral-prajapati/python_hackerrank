```python
# swap case

def swap_case(s):
    new_str = ''
    for i in s:
        if i.islower():
            new_str = new_str+i.upper()
        else:
            new_str = new_str+i.lower()
    return new_str

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

```python
# mutations

def mutate_string(string, position, character):
    lst = list(string)
    lst[position] = character
    new_str = ''.join(lst)
    return new_str

if __name__ == '__main__':
    s = input()
    i, c = input().split()
    s_new = mutate_string(s, int(i), c)
    print(s_new)
```

```python
# find a string

def count_substring(string, sub_string):
    num = []
    lengt = len(string)
    for i in range(lengt):
        if string.find(sub_string,i) == i:
            num.append(i)
    return len(num)

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
```

```python
# text alignment

#Replace all ______ with rjust, ljust or center. 

thickness = int(input()) #This must be an odd number
c = 'H'

#Top Cone
for i in range(thickness):
    print((c*i).rjust(thickness-1)+c+(c*i).ljust(thickness-1))

#Top Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))

#Middle Belt
for i in range((thickness+1)//2):
    print((c*thickness*5).center(thickness*6))    

#Bottom Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))    

#Bottom Cone
for i in range(thickness):
    print(((c*(thickness-i-1)).rjust(thickness)+c+(c*(thickness-i-1)).ljust(thickness)).rjust(thickness*6))
```

```python
# text wrap

import textwrap

def wrap(string, max_width):
    new_str = ''
    x = 1
    for i in string:
        new_str = new_str + i
        
        if x == max_width:
            x = 1
            new_str = new_str + '\n'
            # print(new_str)
        else:
            x += 1
    return new_str

if __name__ == '__main__':
    string, max_width = input(), int(input())
    result = wrap(string, max_width)
    print(result)
```

```python
# string formatting

def print_formatted(number):
    width = len(bin(number)[2:])
    for i in range(1, number + 1):
        decimal = str(i)
        octal = oct(i)[2:]
        hexadecimal = hex(i)[2:].upper()
        binary = bin(i)[2:]
        print(f"{decimal.rjust(width)} {octal.rjust(width)} {hexadecimal.rjust(width)} {binary.rjust(width)}")

if __name__ == '__main__':
    n = int(input())
    print_formatted(n)
```

```python
# alphabet rangoli

def print_rangoli(size):
    alphabet = [chr(ord('a') + i) for i in range(26)]

    width = 4 * size - 3

    for i in range(size):
        left = '-'.join(alphabet[size-j-1] for j in range(i+1))
        row = left + left[-2::-1]
        print(row.center(width, '-'))

    for i in range(size-2, -1, -1):
        left = '-'.join(alphabet[size-j-1] for j in range(i+1))
        row = left + left[-2::-1]
        print(row.center(width, '-'))


if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```

```python
# the minion game

def minion_game(string):
    # your code goes here
    vowels = "AEIOU"
    kevin = sum(len(string)-i for i, char in enumerate(string) if char in vowels)
    stuart = sum(len(string)-i for i, char in enumerate(string) if char not in vowels)
    if kevin == stuart:
        print("Draw")
    elif kevin > stuart:
        print("Kevin", kevin)
    else:
        print("Stuart", stuart)

if __name__ == '__main__':
    s = input()
    minion_game(s)
```

```python
# merge the tools

def merge_the_tools(string, k):
    # your code goes here
    substrings = [string[i:i+k] for i in range(0, len(string), k)]
    
    for s in substrings:
        unique_chars = []
        for c in s:
            if c not in unique_chars:
                unique_chars.append(c)
        print(''.join(unique_chars))

if __name__ == '__main__':
    string, k = input(), int(input())
    merge_the_tools(string, k)
```
