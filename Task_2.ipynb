```python
# find runnerup score

if __name__ == '__main__':
    n = int(input())
    arr = map(int, input().split())
    arr = list(arr)
    
    winner = max(arr)
    runner_up = min(arr)
    
    for i in arr:
        if i != winner and i > runner_up:
            runner_up = i
        
    print(runner_up)
```

```python
# nested lists

if __name__ == '__main__':
    lst = []
    for _ in range(int(input())):
        name = input()
        score = float(input())
        lst.append([score, name])
    lst.sort()
    min_score = lst[0][0]
    while lst != [] and lst[0][0] == min_score:
        lst.pop(0)
    if lst != []:
        second_score = lst[0][0]
        for b in lst:
            if b[0] == second_score:
                print(b[1])
```

```python
# finding the percentage

if __name__ == '__main__':
    n = int(input())
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        scores = list(map(float, line))
        student_marks[name] = scores
    query_name = input()
    
    avg_marks = sum(student_marks[query_name])/len(student_marks[query_name])
    
    print("{0:.2f}".format(avg_marks))
```

```python
# Set .discard(), .remove() & .pop()

n = int(input())
s = set(map(int, input().split()))
num = int(input())

for i in range(num):
    x = list(map(str, input().split()))
    if x[0] == "pop":
        s.pop()
    elif x[0] == "remove":
        try:
            s.remove(int(x[1]))
        except:
            continue
    elif x[0] == "discard":
        try:
            s.discard(int(x[1]))
        except:
            continue
print(sum(s))
```

```python
# set mutation

# Enter your code here. Read input from STDIN. Print output to STDOUT

n = int(input())

set1 = set(map(int, input().split()))

num = int(input())

for i in range(num):
    x = input().split()
    if x[0] == 'update':
        set2 = set(map(int, input().split()))
        set1.update(set2)
    elif x[0] == 'intersection_update':
        set2 = set(map(int, input().split()))
        set1.intersection_update(set2)
    elif x[0] == 'difference_update':
        set2 = set(map(int, input().split()))
        set1.difference_update(set2)
    elif x[0] == 'symmetric_difference_update':
        set2 = set(map(int, input().split()))
        set1.symmetric_difference_update(set2)
    else :
        continue

print(sum(set1))
```

```python
# check strict superset

# Enter your code here. Read input from STDIN. Print output to STDOUT

set1 = set(input().split())
n = int(input())
output = True

for i in range(n):
    set2 = set(input().split())
    
    if not set2.issubset(set1):
        output = False

print(output)
```
